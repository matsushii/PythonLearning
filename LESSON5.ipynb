```python
from sklearn.datasets import make_blobs
import pandas as pd

# ランダムの種3つ、特徴料2つ、塊数2つ、ばらつき1、300個の点
X, y = make_blobs(
    random_state=3,
    n_features=2,
    centers=2,
    cluster_std=1,
    n_samples=300
)
# 特徴料Xのデータフレームを作成
df = pd.DataFrame(X)
# 分類yをtarget列として追加
df["target"] = y
df.head()
```

```python
# グラフをアウトプット行に出力
%matplotlib inline
import matplotlib.pyplot as plt

# 分類によって別々のデータフレームに代入
df0 = df[df["target"]==0]
df1 = df[df["target"]==1]
# 分類0は青、分類1は赤で散布図を描画
plt.figure(figsize=(5,5))
plt.scatter(df0[0], df0[1], color="b", alpha = 0.5)
plt.scatter(df1[0], df1[1], color="r", alpha = 0.5)
plt.grid()
```

```python
X, y = make_blobs(
    random_state=3,
    n_features=2,
    centers=3,
    cluster_std=1,
    n_samples=300
)

df = pd.DataFrame(X)
df["target"] = y

df0 = df[df["target"]==0]
df1 = df[df["target"]==1]
df2 = df[df["target"]==2]

plt.figure(figsize=(5,5))
plt.scatter(df0[0], df0[1], color="b", alpha = 0.5)
plt.scatter(df1[0], df1[1], color="r", alpha = 0.5)
plt.scatter(df2[0], df2[1], color="g", alpha = 0.5)

plt.grid()
plt.show()
```

```python
X, y = make_blobs(
    random_state=3,
    n_features=2,
    centers=5,
    cluster_std=1,
    n_samples=300
)

df = pd.DataFrame(X)
df["target"] = y

df0 = df[df["target"]==0]
df1 = df[df["target"]==1]
df2 = df[df["target"]==2]
df3 = df[df["target"]==3]
df4 = df[df["target"]==4]

plt.figure(figsize=(5,5))
plt.scatter(df0[0], df0[1], color="b", alpha = 0.5)
plt.scatter(df1[0], df1[1], color="r", alpha = 0.5)
plt.scatter(df2[0], df2[1], color="g", alpha = 0.5)
plt.scatter(df3[0], df3[1], color="m", alpha = 0.5)
plt.scatter(df4[0], df4[1], color="c", alpha = 0.5)

plt.grid()
plt.show()
```

```python
from sklearn.datasets import make_regression
# ランダムの種3つ、ノイズ10、特徴料1つ、y切片100、300個の点
X, y = make_regression(
    random_state=3,
    noise = 10,
    n_features=1,
    bias = 100,　
    n_samples=300
)
df = pd.DataFrame(X)
plt.figure(figsize=(5,5))
plt.scatter(df[0], y, color="b", alpha=0.5)
plt.grid()
plt.show()
```
